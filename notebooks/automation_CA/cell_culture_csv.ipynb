```python
import pandas as pd
from datetime import date
```

```python
REP_NUM = 8
COMPOUND_LAYOUT = 1
TX_PLATE_BARCODE = "12345"

CELL_TYPE = "52b2636a-e77c-4574-b6f2-78ad283c489e" # -> hard-coded
CULTURE_STATE = "Active" # -> hard-coded
PASSAGE = 8
PARENT_CULTURE = "fedc28b9-a96f-47ba-a748-c1101fb69d4a"

NUM_CELLS_SEEDED = 400_000
VOL = 0.17

TX_PROTOCOL_ID = "PR003-V1"
PREP_BY = "Yang Gao"
```

```python
treatment_info = pd.read_csv("../../data/Treatment_Layout_for_testing.csv")
treatment_info
```

```python
cell_culture_metadata = pd.DataFrame(
    columns=[
        # "cell_type",  # constant
        "id", 
        # "culture_state",  #constant
        # "passage", #constant
        # "parent_culture", # constant?
        # "num_cells_seeded", # constant
        # "vol(ml)",  # constant?
        "tx_type", 
        "tx_stock_id", 
        "tx_dose", 
        "tx_dose_unit", 
        # "prep_by", # constant
        # "date", # constant
        "well_position"
    ]
)
```

```python
dmso = 'a'
dbet = 'a'

for index, row in treatment_info.iterrows():
    item = None
    if "dmso" in row["Condition"].lower():
        item = dmso
        dmso = chr(ord(dmso) + 1)
    elif "control" in row["Condition"].lower():
        item = dbet
        dbet = chr(ord(dbet) + 1)
    temp_dict = {}
    if row["Condition"].lower() == "dmso":
        name = "THP1_" + "DMSO" + "_rep" + str(REP_NUM)
        temp_dict["id"] = name
        temp_dict["tx_stock_id"] = ""
        temp_dict["tx_dose"] = ""
        temp_dict["tx_dose_unit"] = ""
        
    else:
        name = "THP1_" + row["Compound ID"] + "_rep" + str(REP_NUM)
        temp_dict["id"] = name
        temp_dict["tx_stock_id"] = row["Compound Aliquot ID"]
        temp_dict["tx_dose"] = row["Dose"]
        temp_dict["tx_dose_unit"] = row["Dose Unit"]

    if item:
        temp_dict["id"] = name + item

    temp_dict["tx_type"] = row["Condition"]
    temp_dict["well_position"] = row["Plate Well"]
    print(temp_dict)

    # df2 = {'Name': 'Amy', 'Maths': 89, 'Science': 93}
    cell_culture_metadata = pd.concat([cell_culture_metadata, pd.DataFrame([temp_dict])], ignore_index = True)
```

```python
cell_culture_metadata["cell_type"] = CELL_TYPE
cell_culture_metadata["culture_state"] = CULTURE_STATE
cell_culture_metadata["passage"] = PASSAGE
cell_culture_metadata["parent_culture"] = PARENT_CULTURE
cell_culture_metadata["num_cells_seeded"] = NUM_CELLS_SEEDED
cell_culture_metadata["vol(ml)"] = VOL
cell_culture_metadata["prep_by"] = PREP_BY
cell_culture_metadata["date"] = date.today()
```
